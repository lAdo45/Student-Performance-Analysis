```python
import numpy  as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df=pd.read_csv("student_data.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#drop unnamed column

df = df.drop(columns=["Unnamed: 0"], errors="ignore")
print(df.head())
```

```python
df["WklyStudyHours"]=df["WklyStudyHours"].str.replace("05-oct","5-10")
df.head()
```

```python
#gender distribution

plt.figure(figsize=(5,5))
ax = sns.countplot(data=df, x="Gender" , palette=["blue", "orange"])
ax.bar_label(ax.containers[0])
plt.show()
```

```python
gb= df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb , annot= True)
plt.title("Relationship between parents education and students score")
plt.show()
```

from the above chart we have concluded that the education of the parents have a good impact on their score.


```python
gb1= df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1 , annot= True)
plt.show()
```

from the above chart we have concluded that  there is no/negligible
impact on the student's score due to their parents marital status
